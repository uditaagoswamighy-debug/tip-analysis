```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset("tips")
df.head()
```

```python
df.groupby("time")["tip"].mean()
```

## Tip Analysis Project

**Question:**  
Do customers tip more during lunch or dinner?


**Insight:**

The average tip during dinner (~3.10) is higher than during lunch (~2.73).
This indicates that customers tend to tip more during dinner compared to lunch.


```python
df.groupby("time")["tip"].mean().plot(
    kind="bar",
    title="Average Tip by Time of Day",
    ylabel="Average Tip",
    xlabel="Time of Day"
)
```

## Conclusion

Dinner receives higher average tips than lunch.


```python
df.groupby("smoker")["tip"].mean()
```

**Question:**
Do smokers tip more than non-smokers?

**Insight:**
The average tip from smokers (3.00) is marginally higher than that of non-smokers (2.99).
However, the difference is minimal, suggesting that smoking status alone is unlikely to be a significant driver of tipping behavior.

```python
df.groupby("smoker")["tip"].mean().plot(
    kind = "bar",
    title = "Average Tip by Smokers and Non-Smokers",
    ylabel = "Average Tip",
    xlabel = "Smoker"
)
```

## Conclusion
The analysis shows that smokers have a marginally higher average tip (3.00) compared to non-smokers (2.99).
However, the difference is minimal, suggesting that smoking status alone is unlikely to be a strong factor influencing tipping behavior.
Other variables such as bill size, dining time, or group size may play a more significant role.
